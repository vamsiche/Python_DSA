```python
# statement: You are given a sorted array of integers and a target, your task is to search for the target in the given array.
# Assume the given array does not contain any duplicate numbers.

# Let’s say the given array is = {3, 4, 6, 7, 9, 12, 16, 17} and target = 6.
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while(low<=high):
        mid = (low + high)// 2
        if target == arr[mid]:
            return mid
        elif target > arr[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return -1
    
    
arr = [3,4,6,7,9,12,16,17]
target = 12
binary_search(arr, target)


```

# Recursive way

```python
def binary(arr, low, high, target):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid 
    elif target > arr[mid]:
        return binary(arr, mid+1, high, target)
    else:
        return binary(arr, low, mid-1,target)
    
    
arr = [1,2,3,4,5,6]
target = 5
low = 0
high = len(arr)-1
binary(arr, low, high, target)
```
