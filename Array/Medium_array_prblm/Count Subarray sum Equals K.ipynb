# Brute Force

```python
# Statement: Given an array of integers and an integer k, return the total number of subarrays whose sum equals k.

# A subarray is a contiguous non-empty sequence of elements within an array.

# Pre-requisite: Longest subarray with given sum

# Examples

# Example 1:
# Input Format: N = 4, array[] = {3, 1, 2, 4}, k = 6
# Result: 2
# Explanation: The subarrays that sum up to 6 are [3, 1, 2] and [2, 4].

# Example 2:
# Input Format: N = 3, array[] = {1,2,3}, k = 3
# Result: 2
# Explanation: The subarrays that sum up to 3 are [1, 2], and [3].













def SubArray(arr,k):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i, n):
            Subarray_sum = sum(arr[i:j+1])
            if Subarray_sum == k:
                count+=1            
    return count

arr = [1,1,1]
k = 2
SubArray(arr, k)
            
            
```

# Better approach

```python
def SubArray(arr, k):
    n = len(arr)
    count = 0
    for i in range(n):
        subarray_sum = 0
        for j in range(i, n):
            subarray_sum = subarray_sum + arr[j]
            if subarray_sum == k:
                count+=1
    return count

arr = [1,2,3]
k = 3
SubArray(arr, k)       
```
