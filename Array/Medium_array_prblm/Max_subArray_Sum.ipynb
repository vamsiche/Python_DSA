```python
#  Statement: Given an integer array arr, find the contiguous subarray (containing at least one number) which
# has the largest sum and returns its sum and prints the subarray.

# Examples
# Example 1:

# Input: arr = [-2,1,-3,4,-1,2,1,-5,4] 

# Output: 6 

# Explanation: [4,-1,2,1] has the largest sum = 6. 

# Examples 2: 

# Input: arr = [1] 

# Output: 1 

# Explanation: Array has only one element and which is giving positive sum of 1. 

import sys 
def Max_sub_Array(arr):
    sum = 0
    max = -sys.maxsize - 1
    for i in range(len(arr)):
        sum = sum+arr[i]
        if sum > max:
            max = sum
        if sum <0:
            sum = 0
    return max

arr = [2, -1, 4, -2, 1, -1]
Max_sub_Array(arr)





```

```python
#Brute
import sys
def subarray(arr):
    n = len(arr)
    maxi = -sys.maxsize - 1
    for i in range(n):
        sum = 0
        for j in range(i,n):
            sum = sum + arr[j]
            maxi = max(maxi, sum)
    return maxi

arr = [2, -1, 4, -2, 1, -1]
subarray(arr)
```

# Kadanes Algorithm


```python

def Max_sub_Array(arr):
    sum = 0
    max_sum = -sys.maxsize - 1
    ans_start = -1
    ans_end = -1
    
    for i in range(len(arr)):
        if sum == 0:
            start = i
        sum = sum + arr[i]

        if sum > max_sum:
            max_sum = sum
            ans_start = start
            ans_end = i

        if sum < 0:
            sum  = 0
            start = start - 1

    return max_sum, arr[ans_start:ans_end+1]

arr = [2, -1, 4, -2, 1, -1]
Max_sub_Array(arr)   
    
```
