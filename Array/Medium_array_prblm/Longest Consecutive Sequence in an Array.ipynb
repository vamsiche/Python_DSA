```python
#  Statement: You are given an array of ‘N’ integers. 
# You need to find the length of the longest sequence which contains the consecutive elements.

# Examples

# Example 1:

# Input: [100, 200, 1, 3, 2, 4]

# Output: 4

# Explanation: The longest consecutive subsequence is 1, 2, 3, and 4.

# Input: [3, 8, 5, 7, 6]

# Output: 4

# Explanation: The longest consecutive subsequence is 5, 6, 7, and 8.



def linearSearch(arr, num):
    n = len(arr)  
    for i in range(n):
        if arr[i] == num:
            return True
    return False

def Longest_Consecutive_Sequence(arr):
    n = len(arr)
    long_seq = 1
    for i in range(n):
        x = arr[i]
        count = 1
        while linearSearch(arr, x+1):
            x+=1
            count+=1
        long_seq = max(long_seq, count)
    return long_seq

arr = [3, 8, 5, 7, 6]
Longest_Consecutive_Sequence(arr)
```

```python
def Longest_Consecutive_Sequence(arr):
    n = len(arr)
    if n == 0:
        return 0
    
    longest = 1
    new_set = set(arr)
    for i in new_set:
        if i - 1 not in new_set:
            count  = 1
            x = i
            while x+1 in new_set:
                count+=1
                x+=1
            longest = max(longest, count)
    return longest


arr = [3, 8, 5, 7, 6]
Longest_Consecutive_Sequence(arr)
```
