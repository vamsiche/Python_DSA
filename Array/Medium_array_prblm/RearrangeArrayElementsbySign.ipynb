```python
# Statement:

# There’s an array ‘A’ of size ‘N’ with an equal number of positive and negative elements. 
# Without altering the relative order of positive and negative elements, you must return an array of alternately positive and negative values.

# Note: Start the array with positive elements.

# Examples: 

# Example 1:

# Input:
# arr[] = {1,2,-4,-5}, N = 4
# Output:
# 1 -4 2 -5

# Explanation: 

# Positive elements = 1,2
# Negative elements = -4,-5
# To maintain relative ordering, 1 must occur before 2, and -4 must occur before -5.

# Example 2:
# Input:
# arr[] = {1,2,3,-1,-2,-3}, N = 6
# Output:
# 1 -3 2 -1 3 -2
# Explanation: 

# Positive elements = 1,2,3
# Negative elements = -3,-1,-2
# To maintain relative ordering, 1 must occur before 2, and 2 must occur before 3.
# Also, -3 should come before -1, and -1 should come before -2.

def Re_Arrange_Array_elements_By_sign(arr):
    n = len(arr)
    pos = []
    neg = []
    for i in range(n):
        if arr[i] > 0:
            pos.append(arr[i])
        else:
            neg.append(arr[i])

    for i in range(len(pos)):
        arr[2*i] = pos[i]
    for i in range(len(neg)):
        arr[2*i+1] = neg[i]
    return arr
    


arr = [1,2,3,-1,-2,-3]
Re_Arrange_Array_elements_By_sign(arr)
```
