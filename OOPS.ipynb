## Constructor

```python
# Constructor --> it is method, it runs automatically when object is created

class vehicle:
    def __init__(self):
        pass
    

# Types of Constructors
# 1. Default constructor(Non - Parameterized) 
#                   --> constructor with no parameters except self 

class vehicle:
    def __init__(self):
            pass

# 2. Parameterized Constructor 
#                   --> Constructor that accepts parameters to initialize object data.

class vehicle:
        def __init__(self,name, speed):
            self.name = name
            self.speed = speed

v1 = vehicle("Tesla", 80)
print(v1.name,v1.speed)      

# self --> self represents instance of class, allows access to its variables and methods
# self is not a keyword, but NEVER rename it
# self is instance of class 
```

## Encapsulation

```python

# Encapsulation --> It is the process of binding data (variables) and methods (functions) together into a single unit (class) 
# and restricting direct access to the data to protect it from misuse.

# simple -->  Encapsulation is hiding data and allowing access only through access modifiers.

# Access modifiers such as:

#       Public --> Accessible to everyone               
#                                / represents -- public
                                
#       Private --> Accessible only inside the class   
#                           
#                               / represents -- __private

#       Protected --> Accessible within the class and its subclasses   
#                               / represents -- _protected





class bank:
    def __init__(self):
        self.name = "Vamsi" #public
        self._balance = 500 # protected
        self.__pin = 1234 # private
        
b = bank()
print(b.name)
print(b._bank__pin)
```

## Inheritance


```python
# Inheritance --> child class inherit the properties and methods of parent class

class Parent:
    def show(self):
        print("Parent method")
class Child(Parent):
    pass
```

```python

# single inheritance --> one child class and one parent class
class parent:
    def par(self):
        print("parent class")
        
class child(parent):
    def chi(self):
        print("child class")
        
        
obj = child()
obj.par()
obj.chi()
```
