```python
# Problem Statement: The beauty of a string is defined as the difference between the frequency of the most frequent character
# and the least frequent character (excluding characters that do not appear) in that string.

# Given a string s, return the sum of beauty values of all possible substrings of s.

# Examples
# Example 1:
# Input:
#  s = "xyx"
# Output:
#  1
# Explanation:
#  The substrings with non-zero beauty are:
# "xyx" → frequencies: x:2, y:1 → beauty = 2 - 1 = 1
# "xy" → x:1, y:1 → beauty = 0
# "yx" → y:1, x:1 → beauty = 0
# "x" or "y" → beauty = 0
# Total sum = 1 (from "xyx") = 1

# Example 2:
# Input:
#  s = "aabcbaa"
# Output:
#  17
# Explanation:
#  Various substrings such as "aabc", "bcba", etc., have non-zero beauty values. Summing all gives 17.

def frequency(s):
    total_count = 0
    for i in range(len(s)):
        frequency = {}
        
        for j in range(i, len(s)):
            if s[j] in frequency:
                frequency[s[j]] += 1
            else:
                frequency[s[j]] = 1
            
            values = frequency.values()
            maxi = max(values)
            mini = min(values)
            
            total_count = total_count + (maxi - mini)
    return total_count


s = "aabcbaa"
frequency(s)
```
