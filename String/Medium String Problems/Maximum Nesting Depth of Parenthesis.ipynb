```python
# Statement: Given a valid parentheses string s, return the nesting depth of s. 
# The nesting depth is the maximum number of nested parentheses.

# Examples
                
# Example 1:
# Input: s = "(1+(2*3)+((8)/4))+1"
# Output: 3
# Explanation: Digit 8 is inside of 3 nested parentheses in the string.

# Example 2:
# Input: s = "(1)+((2))+(((3)))"
# Output: 3
# Explanation: Digit 3 is inside of 3 nested parentheses in the string.
            
def nesting_depth(s):
    current_depth = 0
    max_depth = 0
    for char in s:
        if char == "(":
            current_depth += 1
            if current_depth > max_depth:
                max_depth = current_depth
        elif char == ")":
            current_depth -= 1
    return max_depth

s = "(1)+((2))+(((3)))"
nesting_depth(s)
```
