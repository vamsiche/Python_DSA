```python
# Statement: Given a string s, reverse the words of the string.

# Examples:

# Example 1:
# Input: s=”this is an amazing program”
# Output: “program amazing an is this”

# Example 2:
# Input: s=”This is decent”
# Output: “decent is This”


def reverse_words(s):
    words = s.split()
    left = 0
    right = len(words) - 1
    while left < right:
        words[left], words[right] = words[right], words[left]
        left += 1
        right -= 1
    return " ".join(words)


        
    

s = 'this is an amazing program'
reverse_words(s)
```
