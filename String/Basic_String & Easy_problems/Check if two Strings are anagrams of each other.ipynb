```python
#  Statement: Given two strings, check if two strings are anagrams of each other or not.

# Examples
# Example 1:
# Input: CAT, ACT
# Output: true
# Explanation: Since the count of every letter of both strings are equal.

# Example 2:
# Input: RULES, LESRT 
# Output: false
# Explanation: Since the count of U and T  is not equal in both strings.
    
def anagrams(s, t):
    if len(s) != len(t):
        return False
    
    if sorted(s) == sorted(t):
        return True
    return False
s = "CAT"
t = "ACT"
anagrams(s, t)
```
