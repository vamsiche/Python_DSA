```python
# Statement: Given a string s, representing a large integer, the task is to return the largest-valued odd integer (as a string)
# that is a substring of the given string s.
# The number returned should not have leading zero's. But the given input string may have leading zero.

# Examples
                
# Example 1
# Input: s = "5347"
# Output: "5347"
# Explanation: The odd numbers formed by the given string are → 5, 3, 53, 347, 5347. 
# \The largest odd number without leading zeroes is 5347.

# Example 2
# Input: s = "0214638"
# Output: "21463"
# Explanation: The odd numbers formed by the string are → 1, 3, 21, 63, 463, 1463, 21463. 
# ithWe can't use numbers starting with 0, so the largest valid odd number is 21463.


def odd_Num(s):
    ind = 0
    for i in range(len(s)-1,-1,-1):
        if int(s[i])%2==1:
            ind = i
            break
    start = 0
    while(s[start] == "0"):
        start += 1
    return s[start:ind+1]


s = "2146380"
print(odd_Num(s))
            
    
```
